```python
df_filtered = df[['regio', # neighbourhood
                  'gm_naam', # municipality, only for simple reading afterwards
                  'a_inw', #inhabitants
                  'a_vrouw', #women amount
                  'a_00_14', #0-15
                  'a_15_24', #15-24
                  'a_25_44', #25-44
                  'a_45_64', #45-64
                  'a_geb', #births
                  'a_hh_m_k', # households with children
                  'a_ongeh', #unmarried
                  'p_huurw',  # perc. rental 
                  'p_arb_pp', # perc. nett working participaton
                  'a_opl_lg',  # amount low education
                  'a_opl_md',  #amount mid edu
                  'a_opl_hg', # amount high edu
                  'g_hh_sti', # average standardized income
                  'g_pau_hh', #cars per hh
                  'pst_mvp' # most occuring postal
                 ]]


df_renamed = df_filtered.rename(columns={'regio': 'neighbourhood', 
                                         'gm_naam': 'municipality',
                                         'a_inw': 'citizens', 
                                         'a_vrouw': 'females', 
                                         'a_ongeh': 'not_married', 
                                         'p_huurw': 'rental_perc', 
                                         'p_arb_pp': 'percentage_working', 
                                         'a_opl_lg': 'low_edu',  
                                         'a_opl_md': 'mid_edu',  
                                         'a_opl_hg': 'high_edu', 
                                         'a_geb': 'births',
                                         'g_hh_sti': 'avg_income',
                                         'a_hh_m_k': 'households_w_children', 
                                         'g_pau_hh': 'cars_per_household', 
                                         'pst_mvp': 'postal_code_center'
                                        })

df_renamed
```

```python
## some more preprocessing
df_new = df_renamed
df_new.dtypes
```

```python
df_new['percentage_under_15'] = (df['a_00_14'] / df['a_inw']) * 100
df_new
#for column in df_new.columns:
#    if (column.name = 'neighbourhood'):
#        continue
#    df_new[column] = df_new[column] = df_new[column].replace('.', str(0)).str.replace(',', '.').astype(float)
```

```python
df_new['avg_income'] = df_new['avg_income'].replace('.', str(0))
df_new['avg_income'] = df_new['avg_income'].str.replace(',', '.').astype(float)
df_new['cars_per_household'] = df_new['cars_per_household'].replace('.', str(0))
df_new['cars_per_household'] = df_new['cars_per_household'].str.replace(',', '.').astype(float)
df_new['rental_perc'] = df_new['rental_perc'].replace('.', str(0))
df_new['rental_perc'] = df_new['rental_perc'].str.replace(',', '.').astype(float)

df_new
```

```python
# Assign preprocessed dataset to new variable
cbs_data_prepped = df_new

def calculate_demand_values(data):
    # Children under 15
    data['play_child_value']= data['a_00_14']\
        .apply(lambda x: 2 if (x > 800) else (1 if (x>360) else 0 ) )
    # Households with children
    data['play_household_with_children']= data['households_w_children']\
        .apply(lambda x: 2 if (x > 800) else (1 if (x>360) else 0 ) )
    # income level
    data['play_income']= data['avg_income']\
        .apply(lambda x: -1 if (x < 24.2) else 0)
    # cars per household
    data['play_cars']= data['cars_per_household']\
        .apply(lambda x: 1 if (x > 0.5) else 0 )
    # births
    data['play_births']= data['births']\
        .apply(lambda x: 1 if (x > 9) else 0 )
    # rental homes
    data['play_rental']= data['rental_perc']\
        .apply(lambda x: 1 if (x > 41) else (0 if (x>21) else -1 ) )
    # calculated sum of demand
    data['play_demand'] = data['play_child_value']\
                          + data['play_household_with_children']\
                          + data['play_income']\
                          + data['play_cars']\
                          + data['play_births']\
                          + data['play_rental']
    
    return data

demand_included_playground = calculate_demand_values(cbs_data_prepped)
```

```python
def calculate_demand_values_perc(data):
    # Children under 15
    ## POSSIBLE percentage of children
    data['play_percentage_child']= data['percentage_under_15']\
        .apply(lambda x: 2 if (x > 20) else (1 if (x>15) else 0 ) )
    # Households with children
    data['play_household_with_children']= data['households_w_children']\
        .apply(lambda x: 2 if (x > 800) else (1 if (x>360) else 0 ) )
    # income level
    data['play_income']= data['avg_income']\
        .apply(lambda x: -1 if (x < 24.2) else 0)
    # cars per household
    data['play_cars']= data['cars_per_household']\
        .apply(lambda x: 1 if (x > 0.5) else 0 )
    # births
    data['play_births']= data['births']\
        .apply(lambda x: 1 if (x > 9) else 0 )
    # rental homes
    data['play_rental']= data['rental_perc']\
        .apply(lambda x: 1 if (x > 41) else (0 if (x>21) else -1 ) )
    # calculated sum of demand
    data['play_demand'] = data['play_child_value']\
                          + data['play_household_with_children']\
                          + data['play_income']\
                          + data['play_cars']\
                          + data['play_births']\
                          + data['play_rental']
    
    return data

demand_included_playground_perc = calculate_demand_values_perc(cbs_data_prepped)
```
